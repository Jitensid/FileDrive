# 2D Plots 

<p>43. barplot :A bar chart or bar graph is a chart or graph that presents categorical data with rectangular bars with heights or lengths proportional to the values that they represent. The bars can be plotted vertically or horizontally.

 </p>

```python
# Create an fig instance
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

langs = ['Cricket', 'Table Tennis', 'Football', 'BasketBall']
students = [230,1000,350,190]

barplot = ax.bar(langs,students)
#Set X axis label
plt.xlabel("Games")
#Set Y axis label
plt.ylabel("No of Students")
plt.show()
```

```python
N = 3
boys = (5,10,17)
girls = (10,15,20)

ind = np.arange(N) 
# Create an fig instance
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.bar(ind, boys, color='y')
ax.bar(ind, girls,bottom=boys, color='g')

ax.set_title("Bar Plot")
ax.set_ylabel('Scores')
ax.set_xlabel('Teams')
ax.legend(labels=['Boys', 'Girls'])
plt.show()
```

<p>44. Scatter Plot : Scatter plots are used to plot data points on horizontal and vertical axis in the attempt to show how much one variable is affected by another.</p>

```python
feature1 = np.random.rand(30)
feature2 = np.random.rand(30)
target = np.random.rand(30)

fig=plt.figure()
# Create an fig instance

ax=fig.add_axes([0,0,1,1])
ax.scatter(target, feature1, color='g')
ax.scatter(target, feature2, color='b')

ax.set_xlabel('Target')
ax.set_ylabel('Features')
ax.set_title('scatter plot')

plt.show()
```

<p>45. boxplot : A box plot which is also known as a whisker plot displays a summary of a set of data containing the minimum, first quartile, median, third quartile, and maximum. In a box plot, we draw a box from the first quartile to the third quartile. A vertical line goes through the box at the median. The whiskers go from each quartile to the minimum or maximum. </p>

```python
array1 = np.random.normal(100, 20, 20000)
array2 = np.random.normal(80, 40, 20000)
array3 = np.random.normal(90, 50, 20000)
array4 = np.random.normal(70, 65, 20000)

fig = plt.figure()
# Create an fig instance

ax = fig.add_axes([0,0,1,1])

# Create the boxplot
boxplot = ax.boxplot([array1,array2,array3,array4])

#Set X axis label
plt.xlabel("Weight")

#Set Y axis label
plt.ylabel("Height")
plt.show()
```

<p>46. Pie Chart : A Pie Chart can only display one series of data. Pie charts show the size of items (called wedge) in one data series, proportional to the sum of the items. The data points in a pie chart are shown as a percentage of the whole pie.</p>

```python
fig = plt.figure()
# Create an fig instance

ax = fig.add_axes([0,0,1,1])
ax.axis('equal')
countries = ['Sri Lanka', 'Russia','India', 'U.S.A',]
population = [21,40,125,292]
ax.pie(population, labels = countries,autopct='%1.1f%%')
plt.show()
```

# 3D Plots 

<p>49. 3D line plot : plots a 3dimensional line  </p>

```python
# Create an fig instance
fig = plt.figure()

ax = plt.axes(projection='3d')

z = np.arange(1,100,3)

x = z * np.cos(-10 * z)

y = z * np.tan(200 * z)

ax.plot3D(x, y, z, 'purple')
ax.set_title('3D line plot')
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

<p>50. 3D Sactter Points : plots points in the 3 dimensional space   </p>

```python
# Create an fig instance
fig = plt.figure()

ax = plt.axes(projection='3d')

z = np.arange(1,100,2)

x = z * np.cos(-10 * z)

y = z * np.tan(200 * z)

ax.scatter(x, y, z, 'blue')
ax.set_title('3D Scatter Plot')
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```
